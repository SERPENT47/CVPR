```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
def make_spiral_xy(n,i):
      φ = i/16 * math.pi
      r = 6.5 * ((104 - i)/104)
      x = (r * math.cos(φ) * n)/13 + 0.5
      y = (r * math.sin(φ) * n)/13 + 0.5
      plt.scatter(x,y)
      return (x,y)

P = [make_spiral_xy(1,i) for i in range(100)]
N = [make_spiral_xy(-1,i) for i in range(100)]
plt.show()
```
